```python
# Add the Pandas dependency.
import pandas as pd
import os
# i might use os for the file path
```

```python
#initialize variables to hold file path
students_data_to_load = os.path.join('Resources', 'students_complete.csv')
```

```python
#read students data file and store it in pandas DataFrame
students_data_df = pd.read_csv(students_data_to_load)
students_data_df.tail()
```

```python
# create a list with the names of all to students in dataframe using .tolist() METHOD
# we will be 'cleaning' the names with this list
student_names = students_data_df['student_name'].tolist()

# Create a new list and use it for the for loop to iterate through the list.
students_to_fix = []

# Use an if statement to check the length of the name.
# If the name is greater than or equal to "3", add the name to the list.

for name in student_names:
    if len(name.split()) >= 3:
        students_to_fix.append(name)

# Get the length of the students whose names are greater than or equal to "3".
len(students_to_fix)
        
```

```python
# now we have a list of names that contain either prefixes or suffixes or both
# get a list of prefixes from students_to_fix
prefix = []
for name in students_to_fix:
    if len(name.split()[0]) <= 4:
        prefix.append(name.split()[0])
print(prefix)
```

```python
# now get a suffix list. range on suffix is < or equal to 3
suffix = []
for name in students_to_fix:
    if len(name.split()[-1]) <= 3:
        suffix.append(name.split()[-1])
print(suffix)
```

```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
# Iterate through the "prefixes_suffixes" list and replace them with an empty space, "" when it appears in the student's name.
for word in prefixes_suffixes:
    students_data_df["student_name"] = students_data_df["student_name"].str.replace(word,"")
```
